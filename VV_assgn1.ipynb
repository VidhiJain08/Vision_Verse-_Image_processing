** Import matplotlib.pyplot as plt

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xticks([-0.04,-0.02,0.00,0.02,0.04])
ax.set_xlim(-0.06,0.06)
ax.set_yticks([-0.08,-0.04,0.00,0.04,0.08])
ax.set_ylim(-0.12,0.12)
# ax.set_yticks([])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
plt.show()
```

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import numpy as np
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
axes[0].plot([0,100],[0,200],linestyle='--',color='blue')
axes[0].set_xticks([0, 20, 40, 60, 80, 100])
axes[0].set_yticks([0, 50, 150, 200])
axes[0].set_title('First Plot')

x_values = np.linspace(0, 100, 100)
y_values = x_values**2  # Parabola y = x^2
axes[1].plot(x_values,y_values,color='red')
axes[1].set_xticks([0,20,40,60,80,100])
axes[1].set_yticks([0,2000,4000,6000,8000,10000])
axes[1].set_title('Second Plot')

plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
axes[0].plot([0,100],[0,200],linestyle='--',color='blue')
axes[0].set_xticks([0,20,40,60,80,100])
axes[0].set_yticks([0,50,100,150,200])
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('First Plot')

x_values = np.linspace(0, 100, 100)
y_values = x_values**2  # Parabola y = x^2
axes[1].plot(x_values,y_values,color='red')
axes[1].set_xticks([0,20,40,60,80,100])
axes[1].set_yticks([0,2000,4000,6000,8000,10000])
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('Second Plot')

plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
axes[0].plot([0,100],[0,200],linestyle='--',color='blue',linewidth=5)
axes[0].set_xticks([0,20,40,60,80,100])
axes[0].set_yticks([0,50,150,200])
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('First Plot')

x_values = np.linspace(0, 100, 100)
y_values = x_values**2  # Parabola y = x^2
axes[1].plot(x_values,y_values,color='red',linewidth=0.9,linestyle=':')
axes[1].set_xticks([0,20,40,60,80,100])
axes[1].set_yticks([0,2000,4000,6000,8000,10000])
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('Second Plot')

plt.show()
```
