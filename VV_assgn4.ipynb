```python
import cv2
import pandas as pd
import numpy as np
from PIL import Image, ImageFilter
from google.colab.patches import cv2_imshow
```

```python
img1=cv2.imread("/content/lowContrast.png")
img2=cv2.imread("/content/original.jpg")
```

```python
cv2_imshow(img1)
cv2_imshow(img2)
```

```python
img1_gaussian = cv2.GaussianBlur(img1, (3, 3), 1.5)
cv2_imshow(img1_gaussian)
img2_gaussian = cv2.GaussianBlur(img2, (5, 5), 1.5)
cv2_imshow(img2_gaussian)
```

```python
img1_median = cv2.medianBlur(img1, 5)
cv2_imshow(img1_median)
img2_median = cv2.medianBlur(img2, 5)
cv2_imshow(img2_median)
```

```python
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img1_laplacian = cv2.Laplacian(gray_img1, cv2.CV_64F, ksize=3)
img2_laplacian = cv2.Laplacian(gray_img2, cv2.CV_64F, ksize=3)
cv2_imshow(img1_laplacian)
cv2_imshow(img2_laplacian)
```

```python
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img1_bilateral = cv2.bilateralFilter(gray_img1, d=9, sigmaColor=75, sigmaSpace=75)
img2_bilateral = cv2.bilateralFilter(gray_img2, d=9, sigmaColor=75, sigmaSpace=75)
cv2_imshow(img1_bilateral)
cv2_imshow(img2_bilateral)
```

```python
img1_box = cv2.boxFilter(img1, -1, (5, 5))
img2_box = cv2.boxFilter(img2, -1, (5, 5))
cv2_imshow(img1_box)
cv2_imshow(img2_box)
```

```python
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
gray_img1_box = cv2.boxFilter(gray_img1, -1, (5, 5))
gray_img2_box = cv2.boxFilter(gray_img2, -1, (5, 5))
cv2_imshow(gray_img1_box)
cv2_imshow(gray_img2_box)
```

USES:
1. Gaussian- Smoothing and noise reduction.
2. Median- Noise reduction, particularly salt-and-pepper noise.
3. Laplacian- Edge detection and highlighting changes in intensity.
4. Bilateral- Smoothing while preserving edges.
5. Box- Simple and fast smoothing.

Best results by Bilateral since it also preserves edges.
